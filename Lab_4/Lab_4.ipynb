## В блокноті Jupiter виконати завдання з використанням MathPlot з вхідними та вихідними даними.

#### Варіант №9

$\begin{cases}
8.3x_1 + 8.2x_2 = 134,
\\
-10.2x_1 + 15.2x_2 = 102,
\\
14.4x_1 - 7.3x_2 = 141.
\end{cases}$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
equations = ['$8.3x_1 + 8.2x_2 = 134$', '$-10.2x_1 + 15.2x_2 = 102$', '$14.4x_1 - 7.3x_2 = 141$']
x_1 = np.linspace(-45, 30, 1000)
f = [(134 - 8.3 * x_1)/8.2, (102 + 10.2 * x_1)/15.2, (141 - 14.4 * x_1)/(-7.3)]
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(10, 5))

for i, ax_ in enumerate(axes):
    ax_.set(xlim=(-10, 30), ylim=(-30, 50), title=equations[i])
    ax_.grid(linestyle=':', color='green')

axes[0].plot(x_1, f[0], '-.c')
axes[1].plot(x_1, f[1], '--y')
axes[2].plot(x_1, f[2], ':m')

plt.savefig('images/equations_separate.png')
plt.savefig('images/equations_separate.jpg')
plt.savefig('images/equations_separate.svg')

plt.show()
```

```python
matrices = [np.array([[8.3, 8.2],[-10.2, 15.2]]), np.array([[-10.2, 15.2],[14.4, -7.3]]), np.array([[8.3, 8.2],[14.4, -7.3]])]
rhs = [np.array([134, 102]), np.array([102, 141]), np.array([134, 141])]
solutions = [np.linalg.solve(matrices[i], rhs[i]) for i in range(len(matrices))]
```

```python
fig = plt.figure(figsize=(16, 8), dpi=100)
ax = plt.axes()
fig.add_axes(ax)
ax.set(xlim=(-10, 30), ylim=(-30, 40), xlabel='$x_1$', ylabel='$x_2$', title='All lines in one figure')
ax.grid(linestyle='-.', color='green')

ax.plot(x_1, f[0], '-.c', label='f[0]')
ax.plot(x_1, f[1], '--y', label='f[1]')
ax.plot(x_1, f[2], ':m', label='f[2]')

arctans = [math.degrees(math.atan(-8.3 / 8.2)), math.degrees(math.atan(10.2 / 15.2)), math.degrees(math.atan(-14.4 / -7.3))]
locations = [(-5, 22), (0, 5), (5, -12)]

ax.text(*locations[0], equations[0], rotation=arctans[0], rotation_mode='anchor', transform_rotates_text=True)
ax.text(*locations[1], equations[1], rotation=arctans[1], rotation_mode='anchor', transform_rotates_text=True)
ax.text(*locations[2], equations[2], rotation=arctans[2], rotation_mode='anchor', transform_rotates_text=True)

ax.plot(*solutions[0], marker='o', color='r')
ax.plot(*solutions[1], marker='o', color='g')
ax.plot(*solutions[2], marker='o', color='b')

y_1 = np.maximum(f[2], f[0])
y_2 = np.maximum(f[1], y_1)

plt.fill_between(x_1, y_1, y_2, color='grey', alpha=0.5)


ax.annotate(f"The intersection point of\n f[0] and f[1]", ha="center", xy=solutions[0], xytext=(5, 20), arrowprops=dict(arrowstyle="->"))
ax.annotate(f"The intersection point of\n f[1] and f[2]", ha="center", xy=solutions[1], xytext=(20, 25), arrowprops=dict(arrowstyle="->"))
ax.annotate(f"The intersection point of\n f[0] and f[2]", ha="center", xy=solutions[2], xytext=(13, -10), arrowprops=dict(arrowstyle="->"))

ax.legend()

plt.savefig('images/equations_merged.png')
plt.savefig('images/equations_merged.jpg')
plt.savefig('images/equations_merged.svg')


plt.show()
```
